```
# Convert numeric columns from string to float
numeric_cols = ['X', 'Y', 'FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']
for col_name in numeric_cols:
    df = df.withColumn(col_name, col(col_name).cast('float'))

display(df)
```

```
# Example transformations:
# Add a new column 'temperature_celsius' by converting temperature from Fahrenheit to Celsius
df = df.withColumn('temperature_celsius', (df['temp'] - 32) * 5/9)

# Round the 'value' column to 2 decimal places
df = df.withColumn("temperature_celsius", round(df["temp"], 8))

display(df)
```

```

from pyspark.sql.functions import current_date, current_timestamp, date_format

# Filter rows where the month is 'mar' and temperature is above 20 degrees Celsius
filtered_data = df.filter((col('month') == 'mar') & (col('temperature_celsius') > -20))

filtered_data = filtered_data.withColumn("last_modified_date", date_format(current_timestamp(), "MM-dd-yyyy HH:mm:ss"))

display(filtered_data)
filtered_data.count()
```

```
from pyspark.sql.functions import input_file_name
import os

df_single_partition = filtered_data.coalesce(1)

# Define your ADLS storage details
adls_account_name = "adlsfinalstorage"
adls_container_name = "destinationcontainer"

# Define the output path directory in ADLS
output_path_directory = f"abfss://{adls_container_name}@{adls_account_name}.dfs.core.windows.net/temp_output"

# Write the DataFrame to the temporary directory
df_single_partition.write.mode("append").option("header", "true").csv(output_path_directory)

# List the files in the temporary output path directory
files = dbutils.fs.ls(output_path_directory)

# Find the temporary CSV file path
csv_file_path = [f.path for f in files if f.path.endswith(".csv")][0]

# Define the final file path with the fixed file name
fixed_file_path = f"abfss://destinationcontainer@{adls_account_name}.dfs.core.windows.net/destinationdirectory/transformed_forestfire_data.csv"

# Move the temporary CSV file to the fixed file name location
dbutils.fs.mv(csv_file_path, fixed_file_path)

# Optionally, clean up the temporary directory
dbutils.fs.rm(output_path_directory, recurse=True)
```
